```python
import numpy as np
import pandas as pd
import pymc3 as pm
import arviz as az
import matplotlib.pyplot as plt
#1)
# Funcție pentru generarea datelor polinomiale
def generate_data(order, size, beta_sd):
    np.random.seed(42)
    x = np.linspace(0, 1, size)
    y_true = np.polynomial.polynomial.polyval(x, np.random.normal(0, beta_sd, order + 1))
    y_obs = y_true + np.random.normal(0, 0.5, size)
    return x, y_obs

# Parametri
order = 5  # Modificați aici pentru diferite ordine ale polinomului
size = 100  # Modificați aici pentru diferite dimensiuni ale setului de date

# Generare date
x, y = generate_data(order, size, 10)

# Model PyMC3
with pm.Model() as model_p:
    α = pm.Normal('α', mu=0, sigma=1)
    β = pm.Normal('β', mu=0, sigma=10, shape=order + 1)  # Modificați sigma pentru diferite distribuții
    ε = pm.HalfNormal('ε', 5)
    μ = α + pm.math.dot(β, x)
    y_pred = pm.Normal('y_pred', mu=μ, sigma=ε, observed=y)

    # Inferență
    trace_p = pm.sample(2000, tune=1000)

# Reprezentare grafică a curbei
x_new = np.linspace(0, 1, 100)
y_new = trace_p['α'] + np.dot(trace_p['β'], x_new[:, None].T)

plt.figure(figsize=(10, 6))
plt.scatter(x, y, label='Date observate')
plt.plot(x_new, y_new[:, :], color='red', alpha=0.1)
plt.title(f'Model polinomial de ordin {order}')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

# Arviz pentru evaluarea performanței modelului
az.summary(trace_p, hdi_prob=0.95)

# Repetați pentru sd=100 și sd=np.array([10, 0.1, 0.1, 0.1, 0.1])

#sd=100

# Model PyMC3
with pm.Model() as model_p1:
    α = pm.Normal('α', mu=0, sigma=1)
    β = pm.Normal('β', mu=0, sigma=100, shape=order + 1)  # Modificați sigma pentru diferite distribuții
    ε = pm.HalfNormal('ε', 5)
    μ = α + pm.math.dot(β, x)
    y_pred = pm.Normal('y_pred', mu=μ, sigma=ε, observed=y)

    # Inferență
    trace_p1 = pm.sample(2000, tune=1000)

# Reprezentare grafică a curbei
x_new = np.linspace(0, 1, 100)
y_new = trace_p1['α'] + np.dot(trace_p['β'], x_new[:, None].T)

plt.figure(figsize=(10, 6))
plt.scatter(x, y, label='Date observate')
plt.plot(x_new, y_new[:, :], color='red', alpha=0.1)
plt.title(f'Model polinomial de ordin {order}')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

# sd= sd=np.array([10, 0.1, 0.1, 0.1, 0.1])

# Model PyMC3
with pm.Model() as model_p2:
    α = pm.Normal('α', mu=0, sigma=1)
    β = pm.Normal('β', mu=0, sigma=np.array([10, 0.1, 0.1, 0.1, 0.1]), shape=order + 1)  # Modificați sigma pentru diferite distribuții
    ε = pm.HalfNormal('ε', 5)
    μ = α + pm.math.dot(β, x)
    y_pred = pm.Normal('y_pred', mu=μ, sigma=ε, observed=y)

    # Inferență
    trace_p2 = pm.sample(2000, tune=1000)

# Reprezentare grafică a curbei
x_new = np.linspace(0, 1, 100)
y_new = trace_p2['α'] + np.dot(trace_p['β'], x_new[:, None].T)

plt.figure(figsize=(10, 6))
plt.scatter(x, y, label='Date observate')
plt.plot(x_new, y_new[:, :], color='red', alpha=0.1)
plt.title(f'Model polinomial de ordin {order}')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
